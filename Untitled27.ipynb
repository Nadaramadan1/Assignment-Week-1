```python
import cv2
import numpy as np
```

```python
img = cv2.imread('/content/alexander-lunyov-Hdzn_gIdh1o-unsplash.jpg')
```

```python
print("Shape:", img.shape)
print("Data type:", img.dtype)
print("Min pixel:", img.min())
print("Max pixel:", img.max())
print("Mean pixel:", img.mean())
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
print("Grayscale shape:", gray.shape)
```

```python
print("Top-left 5x5 pixel values:\n", gray[0:5, 0:5])
```

```python
from google.colab.patches import cv2_imshow
cv2_imshow(img)
cv2_imshow(gray)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
normalized = gray.astype(np.float32) / 255.0
print("Normalized range:", normalized.min(), "-", normalized.max())
```
